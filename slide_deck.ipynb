## Monthly Trip Length by User Type

> Customers tend to have a greater variance in trip length across the different months than subscribers do. Subscribers remained fairly steady even in the winter, although there was a small decrease during the winter. Meanwhile, customers have a much larger difference in trip length between summer and winter time. During the summer, trips were the longest, while during the winter trips were shortest.

```python
# Monthly usage duration by user type
plt.figure(figsize=(20,5)) # Spread out the data so figure is easier to read
sns.pointplot(data = ford, x='start_time_month', y='duration_min', hue='user_type', dodge=True, linestyles="");
plt.title('Monthly Usage Duration by User Type')
plt.xlabel('Month');
plt.ylabel('Duration (min)');
```

## Trip Length by Day of Week

> Customers tended to have a greater variance in trip length across the different days of the week than subscribers did. Subscribers tended to have similar usage regardless of day, with only a small uptick on the weekends. Customers, on the other hand, tended to greatly increase their trip length on the weekends compared to their daily length. This makes the usage pattern for customers much more U shaped than for subscribers.

```python
order = ['Sun', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat']
sns.barplot(x='start_time_weekday', y='duration_min', hue='user_type', data = ford, order = order)
plt.title('Daily Usage Duration by User Type')
plt.xlabel('Weekday');
plt.ylabel('Duration (min)');
```

## Trip Length by Start Time

> Trip length tended to be more variable for customers than for subscribers. Customers tended to have much greater trip lengths when starting earlier in the day, particularly around 3am, than they did later in the day, however, there was an uptick in the early afternoon. Subscibers on the other hand tended to have more consistent length of trips, however, subscribers still tended to have slightly longer trips in the morning than later in the day.


```python
plt.figure(figsize=(17,5)) # Spread out the data so figure is easier to read
sns.barplot(x='start_time_hour', y='duration_min', hue='user_type', data = ford)
plt.title('Hourly Usage Duration by User Type')
plt.xlabel('Hour');
plt.ylabel('Duration (min)');
```

## Summary

> Customers tended to have greater variance in trip length as a consequence of multiple factors, including time the trip start, day of the week, and month of the year when compared to subscribers. From this, we can infer that subscribers tend to use the bikes for a specific, reoccuring use that occurs on a regular basis. Further, we can infer that customers tend to use the bikes for less frequent uses, and that customers use bikes for reasons that influenced more by time of day, day of week, and month of year than subscribers. This suggests that there are two very different use cases based on customer type for the bikes.
